```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
cols = ['SiO2(wt.%)', 'TiO2(wt.%)', 'Al2O3(wt.%)', 'Cr2O3(wt.%)', 'FeOtot(wt.%)',
        'MnO(wt.%)', 'NiO(wt.%)', 'MgO(wt.%)']
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['axes.labelsize'] = 15
```

```python
labels = [
    '$\mathregular{SiO_2(wt.\%)}$',
    '$\mathregular{TiO_2(wt.\%)}$',
    '$\mathregular{Al_2O_3(wt.\%)}$',
    '$\mathregular{Cr_2O_3(wt.\%)}$',
    '$\mathregular{FeO_{tot}(wt.\%)}$',
    '$\mathregular{MnO(wt.\%)}$',
    '$\mathregular{NiO(wt.\%)}$',
    '$\mathregular{MgO(wt.\%)}$'
]
```

```python
def make_violinplot(df, Srp, filename):
    fig, axes = plt.subplots(2, 4, figsize = (12, 6), tight_layout = True, dpi = 300)
    xticklabels = [Srp + str(j + 1) for j in range(4)]
    
    for i in range(len(cols)):
        row = i // 4
        col = i % 4
        sns.violinplot(data = df, x = 'km_label', y = cols[i], ax = axes[row, col], cut = 0, scale = 'width')
        axes[row, col].set_xticks(ticks = [0, 1, 2 ,3])
        axes[row, col].set_xticklabels(xticklabels)
        axes[row, col].set_xlabel('Cluster')
        axes[row, col].set_ylabel(labels[i])
        
    plt.savefig('./{}_violin_plot'.format(filename))
```

### High Temperature Serpentine

```python
df_H_Srp = pd.read_excel('Ophiolite.xlsx', sheet_name = 'H_Srp_km')
```

```python
make_violinplot(df_H_Srp, 'H_', 'H_Srp')
```

```python
df_H_Srp.km_label.value_counts()
```

### Low Temperature Serpentine

```python
df_L_Srp = pd.read_excel('Ophiolite.xlsx', sheet_name = 'L_Srp_km')
make_violinplot(df_L_Srp, 'L_', 'L_Srp')
```
